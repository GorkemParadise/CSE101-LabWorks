```python
# Library Delay
#    Check the return date of a student's book. "
#    Define a variable days_late = 4.
#    If days_late > 0, print "The return date has passed! A fine will be applied."
#    Otherwise, print "The book was returned on time."'

days_late = int(input("Days Late?"))
if days_late > 0:
    print("The return date has passed! A fine will be applied.")
else:
    print("The book was returned on time.")
```
