```python
#Exercise 1 – Basic Counter

i = 1
while i <= 20:
    if i % 2 == 0:
        print(f"{i}(even)", end=" ")
    else:
        print(f"{i}(odd)", end=" ")
    i += 1
```

```python
#Exercise 2 – Running Total

total = 0
count = 0

while True:
    num = int(input("Writed number (0=exit): "))
    if num == 0:
        break
    total += num
    count += 1

if count > 0:
    average = total / count
    print(f"Total: {total}, Average: {average}")
else:
    print("Pls write only number.")
```

```python
#Exercise 3 – Grade Classifier

while True:
    score = input("Write score (quit for 'exit'): ")
    
    if score == "exit":
        print("You are quitting.")
        break

    if not score.isdigit():
        print("Pls write only 0-100 number.")
        continue

    score = int(score)

    if score > 100 or score < 0:
        print("It must be between 0 and 100.")
    elif score >= 90:
        print("Grade: A")
    elif score >= 80:
        print("Grade: B")
    elif score >= 70:
        print("Grade: C")
    elif score >= 60:
        print("Grade: D")
    else:
        print("Grade: F")
```

```python
#Exercise 4 – Unique Words

sentence = input("Pls write a sentence: ").lower()
words = sentence.split()

hashMap = {}
for word in words:
    hashMap[word] = hashMap.get(word, 0) + 1

unique_words = list(hashMap.keys())

print(f"Total {len(unique_words)} unique words.")
print(sorted(unique_words))
```

```python
#Exercise 5 – Shopping List Helper

list = []

while True:
    print("-----Shoppin List Menu-----")
    print("1. Add Item")
    print("2. Remove Item")
    print("3. View List")
    print("4. Exit")

    choose = input("Chose 1,2,3,4")

    if not choose.isdigit():
        print("Pls write 1 or 2 or 3 or 4")
        continue

    choose = int(choose)
    
    if choose == "exit":
        print("You are quitting.")
        break

    elif choose == 1:
        item = int(input("How many item"))
        list.append(item)
        print(f"{item} added to list")
    elif choose == 2:
        item = int(input("How many item"))
        if item in list:
            list.remove(item)
            print(f"{item} removed to list")
        else:
            print(f"{item} not found")
    elif choose == 3:
        print(f"Current List: {list}")
    elif choose == 4:
        print("You are quitting")
        break
    else:
        print("only chose 1-4")
```

```python
#Exercise 9 – Matrix Diagonal Analyzer

k = 3
matrix = []

for i in range(k):
    row = []
    for j in range(k):
        m = int(input("Enter numbers"))
        row.append(m)
    matrix.append(row)

print(matrix)

primary_list = []
secondary_list = []

for e in range(k):
    primary_list.append(matrix[e][e])
for f in range(k):
    secondary_list.append(matrix[f][k - 1 - f])
print("primary:", primary_list)
print("secondary:", secondary_list)

primary_sum = sum(primary_list)
secondary_sum = sum(secondary_list)
difference = abs(primary_sum - secondary_sum)

print("Primary sum:", primary_sum)
print("Secondary sum:", secondary_sum)
print("Difference:", difference)

result = [abs(primary_list[g] - secondary_list[g]) for g in range(len(primary_list))]
print("Abs matrix difference:", result)
```

```python
#Exercise 10 – Vowel Breakdown

def vowel_breakdown(word):
    vowels = "aeiou"
    for char in word:
        if char in vowels:
            print(char)

vowel_breakdown("hello")
```
