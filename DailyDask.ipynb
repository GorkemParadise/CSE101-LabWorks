```python
def main():
    while True:
        numbers = list(map(int, input("Dizi girin (0 ile bitirin): ").split()))
        
        if numbers[-1] != 0:
            print("Dizi 0 ile bitmelidir.")
            continue
        
        numbers.pop() 
        
        if len(numbers) < 2:
            print("Geçersiz dizi.")
            break
        
        x = numbers[0]
        y = numbers[1]
        
        if y <= x:
            print("Geçersiz dizi.")
            break
        
        gap = y - x
        valid = True
        
        for i in range(2, len(numbers)):
            expected_diff = gap + (i - 1)
            actual_diff = numbers[i] - numbers[i - 1]
            
            if actual_diff != expected_diff:
                valid = False
                break
        
        if not valid:
            print("Geçersiz dizi.")
            break
        
        count = len(numbers)
        print(f"start: {x}, gap: {gap}, count: {count}")
```

```python
if __name__ == "__main__":
    main()
```
