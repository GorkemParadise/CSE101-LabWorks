```python
# Goal: store two pets as JSON ({"name": "...", "age": ...}), load them back, and show ages [2,

''' 
import json
pets = [{"title": "Mochi", "years": 2}, {"title": "Luna", "years": 4}]
with open("pets.json", "w") as f:
f.write(pets) # not JSON
with open("pets.json") as f:
data = f.read()
ages = [p["age"] for p in data] # wrong keys and type
print(ages)
'''

import json
pets = [{"title": "Mochi", "years": 2}, {"title": "Luna", "years": 4}]
with open("pets.json", "w") as f:
    json.dump(pets, f)
with open("pets.json") as f:
    data = json.load(f)
ages = [p["years"] for p in data]
print(ages)
```

```python
# Goal: User.total counts instances; usernames must be non-empty; __repr__ shows User(username='<name>'). Fix the increment, validation, and repr.

''' 
class User:
total = 0
def __init__(self, username):
if username == "":
print("bad") # should stop
User.total + 1 # no effect
self.username = username
def __repr__(self):
return f"User({username})"
u1 = User("sam")
u2 = User("lee")
print("count", User.total)
print(u1, u2)
'''

class User:
    total = 0
    def __init__(self, username):
        if username == "":
            raise ValueError("Username cannot be empty")
        User.total += 1
        self.username = username
    def __repr__(self):
        return f"User(username='{self.username}')"
u1 = User("sam")
u2 = User("lee")
print("count", User.total)
```
