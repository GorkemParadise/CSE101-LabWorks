```python
#Arithmetic Exercise
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
print("Sum:", x + y)
print("Difference:", x - y)
print("Product:", x * y)
print("Division:", x / y)
print("Remainder:", x % y)
print("Power:", x ** y)
```

```python
#Bitwise Exercise
print("AND:", x & y)
print("OR:", x | y)
print("XOR:", x ^ y)
print("Shift left x<<1:", x << 1)
print("Shift right x>>1:", x >> 1)
```

```python
def rectangle_area(L, W):
    A = L * W
    return A

# Example usage
length = 5
width = 3
print("Area of rectangle:", rectangle_area(length, width))
```

```python
#Binary Exercise

a = 5 # binary: 101
b = 3 # binary: 011
print("a & b =", a & b) # AND
print("a | b =", a | b) # OR
print("a ^ b =", a ^ b) # XOR
print("~a =", ~a) # NOT
print("a << 1 =", a << 1) # Shift left
print("a >> 1 =", a >> 1) # Shift right
```
