```python
# 1. Loops & Conditionals (warm-up)

nums = [3, -2, 5, -1, 2, 0, 0, 3]

def w1(nums):
    total = 0
    for n in nums:
        if n > 0: 
            total += n
    return total

w1(nums)
```

```python
# 2. Lists, Sets, and Uniqueness

text = "Apple banana apple Pear"
def w2(text):
    unique = set()
    for w in text.split():
        unique.add(w.lower())
    return sorted(unique)
w2(text) 
```

```python
# 3. Dict Tally (counts)

word = "hello"
def w3(word):
    counts = {}
    for ch in word:
        if ch in counts:
            counts[ch] += 1
        else:
            counts[ch] = 1
    return counts

w3(word)
```

```python
# 4. Functions with Defaults

def clean(name):
    return name.strip()

def greet(name="friend"):
    cleaned = clean(name)
    print(f"Hello, {cleaned}!")

greet(" Ada    ")
```

```python
# 5. Recursion Refresher
kk = [1,2,5, 15]
def sum_list(kk):
    t = 0
    for i in range (len(kk)):
        t += kk[i]
    return t

sum_list(kk)
```

```python
def w5(kk):
    if len(kk) == 0:
        return 0
    else:
        return kk[0] + w5(kk[1:])
    
w5(kk)
```

```python
# 7 JSON Round Trip

import json

movies = [{"title": "Up", "year": 2009}, {"title": "Coco", "year": 2017}]
with open ("movies.json", "w", encoding="utf-8") as f:
    json.dump(movies, f)
with open ("movies.json") as f:
    movies2 = json.loads(f.read()) 
print(movies2)   
```
