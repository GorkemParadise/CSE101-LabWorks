```python
# 2. List Statistics Reporter
values = [10, 20, 30, 40, 50]

def summarize_numbers(values):
    total = 0
    max = values[1]
    min = values[-1]
    for i in values:
        total += i
        if i > max:
            max = i
        if i < min:
            min = i
    average = total / i
    return values, average, total, max, min, i

summarize_numbers(values)
```

```python
# 4. Unique Word Extractor

def unique_words(text):
    text = text.lower()
    words = text.split()
    return set(words)

unique_words("Selam dosjJNRHQMN sdnjn QWJ")
```

```python
# 5. List Manipulation Trio
items = [10, 2, 2, 5, 20, 40]

def swap_ends(items):  
    new_items = items[:]   
    new_items[0], new_items[-1] = new_items[-1], new_items[0]
    return new_items
#[40, 2, 2, 5, 20, 10]

def remove_item(items, target):
    new_items = items[:]  
    if target in new_items:
        new_items.remove(target)
    return new_items
#[10, 2, 5, 20, 40]

def slice_section(items, start, stop):
    list = items[start:stop]
    return list
#2, 5]
```

```python
# 6. Factor Finder Suite

def factors(n):
    result = []
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            result.append(i)
            if i != n // i:     
                result.append(n // i)
    return sorted(result)

factors(16)
```
