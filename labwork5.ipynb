```python
#4. Smallest & Largest Finder

numbers = [1, 2, 3, 4, 5]

max=numbers[0]
min=numbers[0]

for i in numbers:
    if i > max:
        max = i
    if i < min:
        min = i

print(max, min)
```

```python
#7. Phone Book Look-up

contacts = {"Ada": "555-1234", "Ali": "555-9876"}

name = input("Enter a name: ")

if name in contacts:
    print(contacts[name])
else:
    print("Not found")
```

```python
#8. Word Length Mapper

k = input("Enter a sentence: ")

words = k.split()

word_lengths = {}

for word in words:
    word_lengths[word] = len(word)

print(word_lengths)
```

```python
#9. Score Average with Dictionary

lessons = {"Math": 90, "Science": 85, "English": 95}

total_score = sum(lessons.values())
average_score = total_score / len(lessons)

print(average_score)
```

```python
#10. Inventory Counter

stock = {"pen": 4, "notebook": 2}
item = input("Enter an item: ")

if item in stock:
    print(stock[item])
else:
    print("Not found")

cart = {}
item = input("Enter an item: ")

if item in cart:
    cart[item] += 1
else:
    cart[item] = 1
```
