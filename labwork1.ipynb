```python
# Task 4 (10 pts): Decimal to Binary & Hexadecimal
x = int(input("Please write a number:"))

def bin(n):
    binary = ""
    while n>0:
        binary = str(n%2) + binary
        n //= 2
    return binary or "0"
print(bin(x))
```

```python
# Task 5 (15 pts): Bitwise Operations
x = int(input("Please write a number:"))
y = int(input("Please write a number:"))

print("AND:", x & y)
print("OR:", x | y)
print("XOR:", x ^ y)
print("Shift left x<<1:", x << 1)
print("Shift right x>>1:", x >> 1)
```

```python
# Task 7 (30 pts): Writing Functions for Simple Algorithms
import math

def calculate_lcm(lft):
    Lcm = lft * 3
    return Lcm

def rec_area(L, W):
    A = L * W
    return A

def average(A, B):
    avg = (A+B) / 2
    return avg

def root(a, b, c):
    d = math.sqrt(b**2 - 4*a*c)
    x1 = (-b + d) / (2 * a)
    x2 = (-b - d) / (2 * a)
    return (x1, x2)

print("LCM:", calculate_lcm(10))  # Expected Output: 30

print("Rectangle Area:", rec_area(5, 3))  # Expected Output: 15

print("Average:", average(10, 20))  # Expected Output: 15.0

print("Quadratic Roots:", root(1, -3, 2))  # Expected Output: (2.0, 1.0)
```
