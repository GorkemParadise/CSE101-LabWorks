```python
# Task 14 (10 pts): Coffee Discount (Nested if)

is_member = True
order_amount = 120

if is_member == True:
    if order_amount > 100:
        print("You get a 20% discount!")
    else:
        print("You get a 10% discount!")
else:
    print("No discount available.")
```
